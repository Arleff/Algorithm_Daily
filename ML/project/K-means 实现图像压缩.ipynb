```python
import matplotlib.pyplot as plt # plt 用于显示图片
import matplotlib.image as mpimg # mpimg 用于读取图片
from sklearn.cluster import KMeans
import numpy as np
```

```python
# 加载照片进行展示
pixel = mpimg.imread('./data/bird.png')
```

```python
plt.imshow(pixel)
```

```python
pixel = pixel.reshape((128*128 , 3))
```

```python
# 开始聚类  聚类数量为8
kmeans = KMeans(n_clusters=8, random_state=0).fit(pixel)
newPixel = []
for i in kmeans.labels_:
    newPixel.append(list(kmeans.cluster_centers_[i,:]))
```

```python
newPixel = np.array(newPixel)
newPixel = newPixel.reshape((128,128,3))
```

```python
plt.imshow(newPixel)
```

```python
# 开始聚类  聚类数量为16
kmeans = KMeans(n_clusters=16, random_state=0).fit(pixel)
newPixel = []
for i in kmeans.labels_:
    newPixel.append(list(kmeans.cluster_centers_[i,:]))
newPixel = np.array(newPixel)
newPixel = newPixel.reshape((128,128,3))
plt.imshow(newPixel)
```
